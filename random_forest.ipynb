```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("Pumpkin_seeds_Dataset.csv")
data.head(10)
```

```python
vals = data.columns
cols = ["Area","Perimeter","Major_Axis_Length","Minor_Axis_Length","Convex_Are","Equiv_Diameter","Eccentricity","Solidity"," Extent","Roundness","Aspect_Ration","Compactness","class"]
data.columns=cols
data.append(list(vals))
data.head(10)
```

```python
data.describe()
```

```python
data.info()
```

```python
data_ = data[["Area","Perimeter","Major_Axis_Length","Minor_Axis_Length","Convex_Are","Equiv_Diameter","Eccentricity","Solidity"," Extent","Roundness","Aspect_Ration","Compactness"]]
target = data[["class"]]
```

```python
data_
```

```python
target
```

```python
X_train,X_test,y_train,y_test = train_test_split(data_,target,random_state=42,test_size=0.3)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
#finding important features
fimp = model.feature_importances_
print(fimp)
```

```python
colms = np.array(data_.columns)
indxs = []
for i in range(len(fimp)):
    if(fimp[i]>=0.03):
        indxs.append(colms[i])
print(indxs)
```

```python
X = data[indxs]
X_train,X_test,y_train,y_test = train_test_split(X,target,random_state=42,test_size=0.3)
```

```python
#creating another model to increase accuracy of he model using imporrtant features
model2 = RandomForestClassifier()
model2.fit(X_train,y_train)
```

```python
y_pred = model2.predict(X_test)
accuracy_score(y_pred,y_test)
```
